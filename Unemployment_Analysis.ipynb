```python
import pandas as pd

df = pd.read_csv("Unemployment in India.csv")

df.head()
```

```python
df.columns = df.columns.str.strip()
df = df.dropna()

df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
sns.lineplot(data=df, x='Date', y='Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.title("Unemployment Rate Over Time")
plt.show()
```

```python
plt.figure()
sns.barplot(data=df, x='Region', y='Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.title("State-wise Unemployment")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

covid_data = df[df['Date'].dt.year == 2020]

plt.figure()
sns.lineplot(data=covid_data, x='Date', y='Estimated Unemployment Rate (%)')
plt.title("COVID-19 Impact on Unemployment")
plt.show()
```

Unemployment increased during COVID-19 (2020).Some regions show higher unemployment.Trends vary over time
